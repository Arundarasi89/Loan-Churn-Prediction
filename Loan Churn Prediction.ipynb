```python
#Assessing Multicollinearity in the dataset
import statsmodels.api as sma
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif_arr(df1):
    vif = pd.DataFrame()
    df1 = sma.add_constant(df1)
    vif["variables"] = df1.columns
    vif["VIF"] = [variance_inflation_factor(df1.values,i) for i in range(df1.shape[1])]
    return(vif)
```

```python
vif_arr(df1)
```

• OWN_RES column has a high variance inflation factor, which means it has mutlicollinearity

```python
#Remove columns
#OWN_RES - High Multicollinearity
#DURATION_YEAR - repetitive
#NUM_CREDITS - keeping history column instead
#MALE_SINGLE - there are two other features called MALE_DIV and MALE_MAR_or_DIV
#RADIO/TV - not significant enough to predict defaulters
df2 = df1.drop(columns=['OWN_RES','DURATION_YEARS','NUM_CREDITS','MALE_SINGLE','RADIO/TV'])
df2.head()
```

```python
#List of predictive variables
df2.columns
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
X = df2.drop('RESPONSE',axis=1)
Y = df2['RESPONSE']
```

```python
#Split the dataset into 80:20 ratio that is training set is 80% of data and testing set is 20% of data
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2)
```

```python
#Balancing the dataset as earlier we saw that the dataset was imbalanced
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train,y_train)
```

```python
#Create model of Gradient Boosting Classifier
gbc=GradientBoostingClassifier(n_estimators=500,learning_rate=0.05,random_state=100,max_features=5 )
```

```python
#Fit the x train and y train data in the model
gbc.fit(X_res,y_res)
```

```python
#Confusion matrix
print(confusion_matrix(y_test, gbc.predict(X_test)))
```

```python
#Performance Evaluation - Accuracy score
print("GBC accuracy is %2.2f" % accuracy_score(y_test, gbc.predict(X_test)))
```

```python
#Predict Gradient Boosting Classifier
pred=gbc.predict(X_test)
```

```python
#Classification Report
print(classification_report(y_test, pred))
```
